```python
import pandas as pd
```

```python
# Импорт датсета с информаицей о метро
metro = pd.read_csv('metro_data-1488-2025-08-11.csv', delimiter=';')
mtcd = pd.read_csv('mtsd_data-62888-2025-06-17.csv', delimiter=';')
```

```python
# Заполним признак "District" датафрейма mtcd значениями признака "City", если текущее значение - NaN.
mtcd['District'] = mtcd['District'].fillna(mtcd['City'])
# Заполним пробелами признаки "Area" и оставшиеся NaN в "District"
mtcd[['Area', 'District']] = mtcd[['Area', 'District']].fillna('')
```

```python
# Метро, МЦД: Отбор нужных колонок
metro = metro[['Station', 'Line', 'AdmArea', 'District']] 
mtcd = mtcd[['StationName', 'DiameterName', 'District', 'Area']]
mtcd = mtcd.drop(0) # удаление первой строчки (дбликат названия на русском)
metro = metro.drop(0) # удаление первой строчки (дбликат названия на русском)
```

```python
# Переименование колонок mtcd для корректного слияния в одну таблицу с metro
mtcd = mtcd.rename(columns=dict(zip(mtcd.columns, metro.columns)))
```

```python
# Преобразования для единого формата
mtcd['Station'] = mtcd['Station'].str.title()
mtcd['Station'] = mtcd['Station'].str.replace('ё', 'е')
metro['Station'] = metro['Station'].str.title()
metro['Station'] = metro['Station'].str.replace('ё', 'е')
```

```python
# Объединение датафреймов с информацией о станциях метрополитена (metro) и станциях МЦД (mtcd) в один
metro_districts = pd.concat([metro, mtcd], axis=0)
```

```python
metro_districts.head()
```

```python
metro_districts.shape
```

```python
# Преобразование датафрейма metro_districts - если станция метро принадлежит нескольким линиям/ районам, данные группируются в одну строку с разделителем ";"
metro_districts = metro_districts.groupby('Station', as_index=False).agg(lambda x: '; '.join(set(x)))
```
